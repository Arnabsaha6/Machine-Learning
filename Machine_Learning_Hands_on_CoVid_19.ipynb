## Linear regression
![title](https://files.realpython.com/media/fig-lin-reg.a506035b654a.png)


```python
import numpy as np
from sklearn.linear_model import LinearRegression

x_list = [i for i in range(200)]
y_list = [i*2.7 for i in x_list]

print('X: ',x_list[:15])
print('Y: ',y_list[:15])

x = np.array(x_list).reshape((-1, 1))
y = np.array(y_list)
```

```python
model = LinearRegression()
model.fit(x, y)
```

```python
score = model.score(x, y)
print('accuracy of our model is: ', score*100, '%')
```

```python
print(68887*3)
model.predict([[68887]])
```

```python
test_x = [i for i in range(250, 300)]
test_y = [i*3 for i in test_x]

test_X = np.array(test_x).reshape((-1, 1))
predicted_Y = model.predict(test_X)
pd.DataFrame({
    'Predicted': predicted_Y,
    'Original': test_y,
}, index=test_x).plot()
```
